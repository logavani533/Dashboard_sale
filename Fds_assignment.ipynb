8.Check for duplicate records and remove them if present.

```python
data.dtypes
```

```python
data = data.rename(columns={
'Unnamed: 1': 'Order ID',
'Unnamed: 2': 'Date',
'Unnamed: 3': 'Product',
'Unnamed: 4': 'Price',
'Unnamed: 5': 'Quantity',
'Unnamed: 6': 'Purchase Type',
'Unnamed: 7': 'Payment Method',
'Unnamed: 8': 'Manager',
'Unnamed: 9': 'City'
})
# Remove the first row which contains the old headers as data
data = data.iloc[1:].copy()
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')
data['Price'] = pd.to_numeric(data['Price'], errors='coerce')
# Convert 'Quantity' to numeric, then round to nearest integer, then to nullable integer data['Quantity'] = pd.to_numeric(data['Quantity'], errors='coerce').round().astype
data['Product'] = data['Product'].astype('category')
data['Purchase Type'] = data['Purchase Type'].astype('category')
data['Payment Method'] = data['Payment Method'].astype('category')
data['Manager'] = data['Manager'].astype('category')
data['City'] = data['City'].astype('category')
```

```python
data.dtypes
```

9.Create a new column: Total Sales = Quantity × Unit Price (if not already present).

```python
data['Quantity'] = pd.to_numeric(data['Quantity'], errors='coerce').round().astype('Int64')
data['Price'] = pd.to_numeric(data['Price'], errors='coerce')
data['Total_Sales'] = data['Quantity'] * data['Price']
```

```python
print(data[['Quantity', 'Price', 'Total_Sales']].head())
```

10.Extract useful information from date column (year, month, etc.).

```python
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['Day'] = data['Date'].dt.day
data['Weekday'] = data['Date'].dt.day_name()
```

```python
print(data[['Date', 'Year', 'Month', 'Day', 'Weekday']].head())
```

```python
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')
data['Year_Month'] = data['Date'].dt.to_period('M')
monthly_sales = data.groupby('Year_Month')['Total_Sales'].sum()
print(monthly_sales)
```

18. Analyze seasonal trends in sales.

19. Detect any unusual spikes or drops in sales (outliers)

```python
Q1 = data['Total_Sales'].quantile(0.25)
Q3 = data['Total_Sales'].quantile(0.75)
IQR = Q3 - Q1
outliers = data[(data['Total_Sales'] < Q1 - 1.5 * IQR) | (data['Total_Sales'] > Q3 + 1.5 * IQR)]
print(outliers)
```

22. Calculate mean, standard deviation, minimum, and maximum using
NumPy.

```python
sales_array = data['Total_Sales'].to_numpy()
mean_sales = np.mean(sales_array)
std_sales = np.std(sales_array)
min_sales = np.min(sales_array)
max_sales = np.max(sales_array)
```

```python
print(f"Mean: {mean_sales}, Std: {std_sales}, Min: {min_sales}, Max: {max_sales}")
```

23. Compare NumPy results with Pandas results.

```python
mean_np = np.mean(sales_array)
std_np = np.std(sales_array)
min_np = np.min(sales_array)
max_np = np.max(sales_array)
mean_pd = data['Total_Sales'].mean()
std_pd = data['Total_Sales'].std()
min_pd = data['Total_Sales'].min()
max_pd = data['Total_Sales'].max()
```

```python
print("NumPy vs Pandas Statistics:")
print(f"Mean: {mean_np} | {mean_pd}")
print(f"Std: {std_np} | {std_pd}")
print(f"Min:{min_np} | {min_pd}")
print(f"Max:{max_np} | {max_pd}")
```
